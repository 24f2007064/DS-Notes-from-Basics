```python
import json
```

```python
# loading Data
def load_data(filename):
    with open(filename, "r") as f:
        json_data = json.load(f)
    return json_data
```

```python
data = load_data("data.json")
```

```python
# display user data-connection
def user_connection(social_data):
    for user in social_data['users']:
        print(f"id. {user['id']} user: {user['name']} - Friends: {user['friends']} - Liked Pages: {user['liked_pages']}") 
    for page in social_data['pages']:
        print(f"{page['id']}: {page['name']}")
```

```python
data = load_data("data.json")
user_connection(data)
```
