## INTRO in MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Great Sachin's Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years, kohli, label = "Virat")
plt.plot(years, sehwag, label = "Sehwag")

plt.legend()

plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Performance")
```

```python

plt.plot(years, kohli)
plt.plot(years, sehwag)
# BAD WAy
plt.legend(["Kohli", "Sehwag"])

plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Performance")
```

### Format String

```python
plt.plot(years, kohli,'r^--', label = "Virat")
plt.plot(years, sehwag,"bo:", label = "Sehwag")

plt.legend()

plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Performance")
```

### Color & Line argument

```python
plt.plot(years, kohli,color = "gold", linestyle = "--", label = "Virat")
plt.plot(years, sehwag,color = "red", linestyle = ":", label = "Sehwag")
plt.plot(years, runs, color = "blue",linestyle = "--", label = "Sachin")

plt.legend()

plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Performance")

plt.grid(True)
plt.style.use('bmh')
```

```python
import numpy as np

for i in range(50):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
plt.show()
```
