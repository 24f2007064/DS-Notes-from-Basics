## Data Frames
#### 2D Labeled Table

```python
data = {
    "Name" : ["Alice", "Maga", "Bob"],
    "age" : [12, 23, 41],
    "city" : ["Delhi", "Mumbai", "Pune"]}

df = pd.DataFrame(data)
df
```

```python
data = {
    "Name" : ["Alice", "Maga", "Bob"],
    "age" : [12, 23, 41],
    "city" : ["Delhi", "Mumbai", "Pune"]}

df = pd.DataFrame(data, index = ['As', "Hila", "Kima"])
df
```
