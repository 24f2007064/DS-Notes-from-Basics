## Data Frames

### From Python Lists

```python
import pandas as pd
data = [
    ["Ani", 19],
    ['Balis', 20],
    ['Chula', 69]]
data
```

```python
df = pd.DataFrame(data, columns = ["Name", "Age"])
df
```

### From Dictionary of Lists

#### Each key becomes a column, and each list is the column data.

```python
data = {
    "Name" : ['Ani', "Free", "Fire"],
    "Age" : [19, 7, 15]}

data
```

```python
df = pd.DataFrame(data)
df
```

### From CSV Files

```python
df = pd.read_csv("data.csv")
df
```

#### Same we can do for JSON, xlsx by using -> read_json, read_xlsx
