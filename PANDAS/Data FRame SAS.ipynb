```python
import pandas as pd

data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Name"] # Single column (as Series)
```

```python
df.loc[0]    # First row (by label)
```

## Conditional Selection

```python
df[df["Age"]>30]
# Those whose Age is GREATER THAN 30
```

```python
df[df["Age"]>30]["Name"]
```

```python
df[(df["Age"]>30) & (df['City'] == "Paris")]   #AND
# Those with age > 30 AND City is Paris
```

```python
df[(df["Age"]>30) | (df['City'] == "Paris")]   # OR
# Those with Age > 30 or Live in PAris
```
