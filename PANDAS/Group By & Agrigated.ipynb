```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby("Category")

for i, w in cat :
    print(i)
    print(w)
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Store")["Sales"].sum()
```

```python
df.groupby(["Store", "Category"])["Sales"].mean()
```

```python
df["Sales"].agg(["sum", "mean", "max", "median"])
```

```python
df.groupby(["Category", "Store"])["Sales"].agg(
    afg_sales = 'mean',
    median_sales = "median"
)
```
