### Changing Data Types

```python
import numpy as np
arr = np.array([1.3,2.5,3.1,4.7,5.6])
print(arr.dtype)
arr
```

```python
arr_int = arr.astype(np.int64)
print(arr_int)
print(arr_int.dtype)
```

### Downcasting to Save Memory

```python
arr_large = np.array([1000000, 2000000, 3000000], dtype=np.int64)
arr_small = arr_large.astype(np.int32)  # Downcasting to a smaller dtype
print(arr_small)  # Output: [1000000 2000000 3000000]
print(arr_small.dtype)  # Output: int32
```

## Memory Usage

```python
arr64 = np.array([1,2,3,4], dtype = np.int64)
arr32 = np.array([1,2,3,4], dtype = np.int32)

print(arr64.nbytes)
print(arr32.nbytes)
```

## String Array

```python
arr = np.array(['apple', 'banana', 'cherry'], dtype='U10')  # Unicode string array
print(arr)
print(arr.dtype)
```

## Complex Numbers

```python
arr = np.array([1 + 2j, 3 + 4j, 5 + 6j], dtype='complex128')
print(arr)
```

## Object Data Type

```python
arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arr)
print(arr.dtype)
```
